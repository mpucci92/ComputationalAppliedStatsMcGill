```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
credit_data = pd.read_csv("C:\\Users\\mpucci\\Desktop\\Credit.csv",
                          usecols = ["Income","Limit","Rating","Cards","Age","Education","Gender","Student","Married",
                                    "Ethnicity","Balance"])
```

```python
print(credit_data.isna().sum()) #No NaN Values in DataSet
print (credit_data.shape) #400 rows with 11 columns
```

```python
X = credit_data[["Income","Limit","Rating","Cards","Age","Education"]] #Attribute Matrix 
```

```python
credit_data["Balance_Boolean"] = credit_data["Balance"]<=1500
```

```python
credit_data['Balance_factor'] = credit_data["Balance_Boolean"].factorize()[0]
print(credit_data)
credit_data1 = credit_data.copy()
```

```python
credit_data.drop(['Balance_Boolean'],axis=1)
```

```python
counter = 0 
for balance in credit_data.Balance_factor:
    if balance == 1: 
        counter+=1
print ((counter/400)*100) # percentage of trues in dataset. 
```

```python
print("This is the Logisitic Regression Fit")

X = credit_data[['Income','Limit', 'Rating', 'Cards', 'Age', 'Education']]
y = credit_data['Balance_factor']
lr = LogisticRegression()
lr.fit(X, y)

print(lr.intercept_)
print(lr.coef_)
```

```python
print("This is the LDA fit equation")

X = credit_data[['Income','Limit', 'Rating', 'Cards', 'Age', 'Education']]
y = credit_data['Balance_factor']

lda = LinearDiscriminantAnalysis()
lda.fit(X,y)

print(lda.intercept_)
print(lda.coef_)
```

```python
print("This is the Quadratic Fit")

X = credit_data[['Income','Limit', 'Rating', 'Cards', 'Age', 'Education']]
y = credit_data['Balance_factor']

qda = QuadraticDiscriminantAnalysis()
qda.fit(X,y)
```

```python
print("Logistic regression equation and prediction")

X1_pred = np.array([63,8100,600,4,30,14]).reshape(-1,6)

print(lr.predict_proba(X1_pred))

X2_pred = np.array([186,13414,950,2,41,13]).reshape(-1,6)

print(lr.predict_proba(X2_pred))

#x1= 'Income' = 63, 'Limit' = 8100, 'Rating' = 600, 'Cards' = 4, 'Age' = 30, 'Education' =14

#x2= ['Income' = 186, 'Limit' = 13414, 'Rating' = 950, 'Cards' = 2, 'Age' = 41, 'Education' =13
```

```python
print("LDA fit equation and prediction")

X1_pred = np.array([63,8100,600,4,30,14]).reshape(-1,6)

print(lda.predict_proba(X1_pred))

X2_pred = np.array([186,13414,950,2,41,13]).reshape(-1,6)

print(lda.predict_proba(X2_pred))
```

```python
print("QDA fit equation and prediction")
X1_pred = np.array([63,8100,600,4,30,14]).reshape(-1,6)

print(qda.predict_proba(X1_pred))

X2_pred = np.array([186,13414,950,2,41,13]).reshape(-1,6)

print(qda.predict_proba(X2_pred))
```
