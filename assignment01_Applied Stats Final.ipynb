```python
#Load the The data 
import numpy as np
import pandas as pd

income_data = pd.read_csv("C:\\Users\\mpucci\\Desktop\\Income2.csv",usecols= ['Education','Seniority','Income'])
print (income_data)
print(income_data.shape)
```

## 1.	Explore the data through various figures.


```python
#CODE HERE

#1 Descriptive statistics
descriptive_data = income_data.describe()
print (descriptive_data)

#2 Scatter plot, histogram visualization of the data. 
from pandas.plotting import scatter_matrix
scatter = scatter_matrix(income_data)
```

## 2.	Predict “y = Income” as a linear function of “x = Years of Education” using simple linear regression, and predict Income for a new individual with “x = 18”.


```python
#CODE HERE
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X = income_data[['Education']], y = income_data['Income'])

print ("This is the slope of the linear regression",lr.coef_)
print("This is the intercept of the linear regression",lr.intercept_)
print ("This is the regression rule: y=6.38x-41.92")

# lr.predict()

print ("This is the income for 18 years of education",lr.predict(18))

# y = 6.38x-41.92 approximately 

import matplotlib.pyplot as plt

%matplotlib inline
plt.plot(income_data["Education"], income_data["Income"], 'or', mfc='none')

# Regression Line
plt.plot(income_data["Education"], lr.intercept_+lr.coef_*income_data["Education"], "-b")
plt.xlabel('Education (Years)')
plt.ylabel('Income (Dollars)')
```


## 3.	Show the scatter plot of Income versus Years of Education, and add the fitted line over the scatter plot.   


```python
#CODE HERE
import matplotlib.pyplot as plt

%matplotlib inline
plt.plot(income_data["Education"], income_data["Income"], 'or', mfc='none')

# Regression Line
plt.plot(income_data["Education"], lr.intercept_+lr.coef_*income_data["Education"])
plt.title("Income in relation to Years of Education")
plt.xlabel('Education (Years)')
plt.ylabel('Income (Dollars)')
```

## 4.	Predict “y = Income” as a linear function of “x1 = Years of Education” and “x2 = Seniority” for a new individual with (x1 = 18, x2 = 60)


```python
#CODE HERE

lr.fit(X = income_data[ ['Education', 'Seniority'] ], y = income_data['Income'])
x = np.array([18, 60] )
# sample in row, features in column
lr.predict(x.reshape(1,2))

print ("The income for 18 years of education and 60 years of seniority is", " ",lr.predict(x.reshape(1,2)), "dollars")
```
